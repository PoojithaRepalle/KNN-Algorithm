```python
import numpy as np
import random
import matplotlib.pyplot as plt
from collections import Counter

def load_flower_data(file_path):
    flower_data = []
    with open(file_path, 'r') as data_file:
        next(data_file)
        for entry in data_file:
            if entry.strip():
                parts = entry.strip().split(',')
                features = [float(feature) for feature in parts[1:-1]]
                species = parts[-1]
                flower_data.append(features + [species])
    return flower_data

def calc_euclidean_dist(v1, v2):
    return np.sqrt(np.sum((v1 - v2) ** 2))

def knn_predictor(train_set, query_instance, neighbors_count):
    all_distances = []

    for flower in train_set:
        dist = calc_euclidean_dist(np.array(flower[:-1]), np.array(query_instance[:-1]))
        all_distances.append((dist, flower[-1]))

    all_distances.sort(key=lambda x: x[0])
    k_nearest_neighbors = all_distances[:neighbors_count]

    predicted_labels = [neighbor[1] for neighbor in k_nearest_neighbors]
    final_prediction = Counter(predicted_labels).most_common(1)[0][0]

    return final_prediction

def split_data(data, split_ratio=0.2):
    random.shuffle(data)
    split_idx = int(len(data) * (1 - split_ratio))
    return data[:split_idx], data[split_idx:]

def compute_accuracy(true_labels, predicted_labels):
    correct_predictions = np.sum(np.array(true_labels) == np.array(predicted_labels))
    return correct_predictions / len(true_labels)

def build_confusion_matrix(true_labels, predicted_labels, class_labels):
    conf_matrix = np.zeros((len(class_labels), len(class_labels)), dtype=int)
    label_map = {label: idx for idx, label in enumerate(class_labels)}
    for true, pred in zip(true_labels, predicted_labels):
        conf_matrix[label_map[true]][label_map[pred]] += 1
    return conf_matrix

def evaluate_knn(flower_data, k_choices, test_ratio=0.2):
    training_data, testing_data = split_data(flower_data, test_ratio)
    test_features = [sample[:-1] for sample in testing_data]
    test_labels = [sample[-1] for sample in testing_data]

    performance = []
    for k in k_choices:
        test_predictions = []
        for test_case in testing_data:
            predicted_species = knn_predictor(training_data, test_case, k)
            test_predictions.append(predicted_species)

        accuracy = compute_accuracy(test_labels, test_predictions)
        performance.append(accuracy)

        if k == max(k_choices):
            print(f"Accuracy for k={k}: {accuracy * 100:.2f}%")
            print("Confusion Matrix:")
            print(build_confusion_matrix(test_labels, test_predictions, list(set(test_labels))))

    return performance

def plot_performance_graph(k_choices, performance):
    plt.figure(figsize=(8, 6))
    plt.plot(k_choices, performance, marker='o', linestyle='-', color='g')
    plt.title('Neighbors (K) vs Accuracy')
    plt.xlabel('K (Number of Neighbors)')
    plt.ylabel('Accuracy')
    plt.grid(True)
    plt.show()

iris_data = load_flower_data('iris.csv')

k_range = list(range(1, 10))

performance_results = evaluate_knn(iris_data, k_range)

plot_performance_graph(k_range, performance_results)
```
